```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu_1 = 1
sigma_1 = .2
dist_1 = np.random.normal(mu_1, sigma_1, 1000)

fig, ax = plt.subplots(1, 2, figsize = (16, 6))
ax[0].stem(dist_1)
bins = 30
ax[1].hist(dist_1, bins, density = False)
ax[1].plot(bins, 1/(sigma_1 * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu_1)**2 / (2 * sigma_1**2)), linewidth = 20, color='r' )
plt.show()
```

```python
mu_1 = 1
sigma_1 = .2
dist_1 = np.random.normal(mu_1, sigma_1, 1000)

fig, ax = plt.subplots(1, 2, figsize = (16, 6))
ax[0].stem(dist_1)
bins = 30
ax[1].hist(dist_1)
plt.show()
```

```python
m1 = np.mean(dist_1)
std1 = np.std(dist_1)
print(m1, std1)

dist_2 = dist_1 ** 2
m2 = np.mean(dist_2)
std2 = np.std(dist_2)
print(m2, std2)
```

```python
fig, ax = plt.subplots(2, 2, figsize = (16, 8))

ax[0,0].stem(dist_1)
ax[0,1].hist(dist_1)
ax[1,0].stem(dist_2)
ax[1,1].hist(dist_2)

plt.show()
```
